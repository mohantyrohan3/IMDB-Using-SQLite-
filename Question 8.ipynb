# Print the genre and percentage of movies total earning present in that genre with 2 decimal place after rounding off.
# Note: Movies has multiple genres, so calculate the percentage for each genre.Total Earning=Domestic earning + WorldWide earning.

```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import sqlite3
import matplotlib.pyplot as plt
db=sqlite3.connect("IMDB.sqlite")
cur=db.cursor()
```

```python
data=pd.read_sql_query("select genre,Domestic,Worldwide from genre join earning on genre.Movie_id=earning.Movie_id",db)
data.head()
df=data.copy()
# def f(bud):
#     if bud=="":
#         return 0
#     return float(bud)
# df["Budget"]=df["Budget"].apply(f)
# df["genre"].value_counts()
df["Total"]=df["Domestic"]+df["Worldwide"]
df.head()
```

```python
Genre=list(df["genre"].unique())
Genre1=[]
for i in range(len(Genre)):
    if Genre[i]!="":
        Genre1.append(Genre[i])
Genre1
```

```python
values=[]
for i in range(len(Genre)):
    if Genre[i]!="":
        final_df=df[df["genre"]==Genre[i]]
        values.append(final_df["Total"].sum())
values
```
