# Find the Movie with Longest-Running Time from IMDB Movie database.
# Print the name of movie and running time as integer value.

```python
import sqlite3
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'

db=sqlite3.connect("IMDB.sqlite")
cur=db.cursor()
```

```python
df=pd.read_sql_query("select * from IMDB",db)
df.head()
```

```python
final_df=df[df["Runtime"]!=""]
final_df
def f(runtime):
    i=0
    while runtime[i]!=" ":
        i+=1
    ans=runtime[0:i]
    return float(ans)
final_df["Runtime"]=final_df["Runtime"].apply(f)
final_df
```

```python
# Answer  
ans=final_df['Runtime'].max()
final_df[final_df["Runtime"]==ans]
```
