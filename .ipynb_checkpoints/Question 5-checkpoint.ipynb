# Find the highest-grossing movie (i.e domestic earning + worldwide earning) in IMDB database year wise.
# Print the Year and name of the movie which was the highest grosser in that year.
# Note:Print the year in ascending order.

```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import sqlite3
db=sqlite3.connect("IMDB.sqlite")
cur=db.cursor()
```

```python
data=pd.read_sql_query("select Title,Rating,Budget,Domestic,Worldwide from IMDB join earning on IMDB.Movie_id=earning.Movie_id",db)
data.head()
df=data.copy()
df
```

```python
df["Gross"]=df["Domestic"]+df["Worldwide"]
df
```

```python
# making a year column
def f(title):
    i=0
    while i!=len(title):
        if title[i]=='(':
            one=i
        if title[i]==')':
            two=i
        i+=1
    return int(title[one+1:two])

df["Year"]=df["Title"].apply(f)
df
```

```python
df["Year"].value_counts()
#year=[2010,2011,2012,2013,2014,2015,2016]
```

```python
# For Year 2010
year1=df[df["Year"]==2010]
ans=year1["Gross"].max()
year1[year1["Gross"]==ans]
# Toy Story 3 (2010) 2010
```
