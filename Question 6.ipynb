# Find the movie which has genre of 'Sci-Fi' and 'Mystery' and movie rating is greater or equal to 8 from IMDB Movie Database.
# Print the movie name.
# Movies has multiple genre.

```python
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import sqlite3
db=sqlite3.connect("IMDB.sqlite")
cur=db.cursor()
```

```python
data=pd.read_sql_query("select Title,Rating,genre from IMDB join genre on IMDB.Movie_id=genre.Movie_id",db)
data.head()
df=data.copy()
df
```

```python
# Answer
ans1=df[(df["genre"]=="Mystery") | (df["genre"]=="Sci-Fi") & (df["Rating"]>=8)]
ans1
# Arrival (2016)
```
